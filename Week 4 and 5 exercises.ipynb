```r
peanut_lines<-read.csv('peanut_lines.csv')
peanut_disease<-read.table('peanut_disease.txt', sep="\t", header=TRUE)
head(peanut_disease)

unique(peanut_disease$Rating)
```

```r
colnames(peanut_lines)
colnames(peanut_disease)
```

```r
disease_data<-merge(peanut_lines, peanut_disease,  all = TRUE)
```

```r
leaf.spot<-subset(disease_data, subset=Rating=='LS')
head(leaf.spot)
```

```r
leaf.spot<-leaf.spot[!is.na(leaf.spot$Disease),]
leaf.spot
```

```r
mean(leaf.spot$Disease[leaf.spot$Year=="2017"])
```

```r
leaf.spot2<-aggregate(leaf.spot$Disease, list(leaf.spot$NC_Accession),drop = TRUE,mean)
leaf.spot2

#vA-C92R is the most leaf spot resistant
```
